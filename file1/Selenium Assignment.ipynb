```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
driver = webdriver.Chrome()
# Open the URL
driver.get('https://www.naukri.com/')
time.sleep(5)

#search for 'Data Scientist'
search = driver.find_element(By.CLASS_NAME, 'suggestor-input')
search.send_keys("Data Scientist")
search_button = driver.find_element(By.CLASS_NAME, 'qsbSubmit')
search_button.click()
time.sleep(5)


def apply_filter(filter_text):
    retries = 3
    while retries > 0:
        try:
            filters = driver.find_elements(By.CLASS_NAME, 'styles_ellipsis__cvWP1.styles_filterLabel__jRP04')
            for f in filters:
                if f.text == filter_text:
                    f.click()
                    time.sleep(5) 
                    return
        except StaleElementReferenceException:
            print(f"Stale element encountered while applying filter '{filter_text}', retrying...")
            retries -= 1

# Apply the 'Delhi / NCR' location filter
apply_filter('Delhi / NCR')

# Apply the '3-6 Lakhs' salary filter
apply_filter('3-6 Lakhs')

# Initialize lists to store job details
job_titles = []
company_name = []
exp_req = []

# Function to fetch job details
def fetch_job_details(job):
    try:
        title = job.find_element(By.CLASS_NAME, 'row1').text
#         print(title)
        job_titles.append(title)

        try:
            company = job.find_element(By.CLASS_NAME, 'comp-name').text
        except:
            company = 'N/A'
        company_name.append(company)
#         print(company)

        try:
            exp = job.find_element(By.CLASS_NAME, 'ni-job-tuple-icon.ni-job-tuple-icon-srp-experience.exp').text
        except:
            exp = "Not mentioned"
        exp_req.append(exp)
#         print(exp)

    except Exception as e:
        print(f"Error fetching job details: {str(e)}")

# Fetch job details
jobs = driver.find_elements(By.CLASS_NAME, "srp-jobtuple-wrapper")[:10]

for i, job in enumerate(jobs):
    try:
        fetch_job_details(job)
    except StaleElementReferenceException:
        print(f"Stale element encountered on job {i}, re-fetching jobs.")
        jobs = driver.find_elements(By.CLASS_NAME, "srp-jobtuple-wrapper")[:10]  
        fetch_job_details(jobs[i])

# DataFrame 
df = pd.DataFrame({
    'Job Title': job_titles,
    'Company Name': company_name,
    'Experience Required': exp_req
})
print(df)
df.to_csv('naukri_job_listings.csv', index=False)
# driver.quit()
```

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time

driver = webdriver.Chrome()
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART/')
time.sleep(5) 
#Empty Lists
ratings=[]
summary=[]
fulls=[]

def get_reviews():
    try:
        all_reviews = driver.find_elements(By.CLASS_NAME, 'EKFha-')
        for review in all_reviews:
            time.sleep(2)  
            try:
                rating = review.find_element(By.CLASS_NAME, 'XQDdHH.Ga3i8K').text
                ratings.append(rating)
                review_summary = review.find_element(By.CLASS_NAME, 'z9E0IG').text
                summary.append(review_summary)
                full_review = review.find_element(By.CLASS_NAME, 'ZmyHeo').text
                fulls.append(full_review)
#                 print(f"Rating: {rating}, Summary: {review_summary}, Review: {full_review}")
            except NoSuchElementException:
                continue 
    except StaleElementReferenceException:
        print("In the except block for stale element")

def click_next_button():
    try:
        # Wait for the "Next" button to be visible and clickable
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, '_9QVEpD')))
        nav_buttons = driver.find_elements(By.CLASS_NAME, '_9QVEpD')
        
        for nav_button in nav_buttons:
            if nav_button.text == 'NEXT':
#                 print("Clicking the Next button")
                nav_button.click()
                return True
        return False  
    except (StaleElementReferenceException, NoSuchElementException) as e:
        print(f"Error with 'next' button: {str(e)}")
        return False

# Scraping multiple pages
for page in range(10): 
    get_reviews()
    
    if not click_next_button():
        print("Next button not found, stopping.")
        break
    
    # Wait for the next page of reviews to load
    time.sleep(5)

# driver.quit()
df=pd.DataFrame({
    'Rating':ratings,
    'Summary':summary,
    'Full Review':fulls    
})
print(df)

df.to_csv("flipkart iphone 11 reviews.csv")
```

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementClickInterceptedException
import time

# WebDriver
driver = webdriver.Chrome()
driver.get('https://www.flipkart.com/')
time.sleep(3)

# empty lists
brands = []
prices = []
descriptions = []

# search for Sneakers
search = driver.find_element(By.CLASS_NAME, 'Pke_EE')
search.send_keys('Sneakers')
search.send_keys(Keys.ENTER)

def click_next_button():
    try:
        time.sleep(5)  # allow time for elements to load
        # wait for the "Next" button to be visible
        next_buttons = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, '//a[contains(@class, "_9QVEpD")]'))
        )
        
        # identify the "Next" button by its text
        for button in next_buttons:
            if "NEXT" in button.text.upper():
                #visibility of next button
                driver.execute_script("arguments[0].scrollIntoView(true);", button)
                time.sleep(1)  # Allow time for scrolling
                #clicking
                driver.execute_script("arguments[0].click();", button)
                return True
        return False
    except ElementClickInterceptedException:
        print("Element Click Intercepted, retrying...")
        return False
    except (StaleElementReferenceException, NoSuchElementException) as e:
        print(f"Error with 'next' button: {str(e)}")
        return False

def get_product_details():
    # all items
    allitems = driver.find_elements(By.CLASS_NAME, '_1sdMkc.LFEi7Z')
    for item in allitems:
        if len(brands) >= 100:
            return False  # stop when 100 items
        
        try:
            # Extract brand, price, and description
            brand = item.find_element(By.CLASS_NAME, 'syl9yP').text
#             print(brand)
            price = item.find_element(By.CLASS_NAME, 'Nx9bqj').text
#             print(price)
            description = item.find_element(By.CLASS_NAME, 'WKTcLC').text
#             print(description)
            
            # Append the details to the lists
            brands.append(brand)
            prices.append(price)
            descriptions.append(description)
        except StaleElementReferenceException:
            print("Stale element, refetching...")
            continue
    return True

def scrape_pages(pages_to_scrape=3):
    for i in range(pages_to_scrape):
        if not get_product_details():
            print("Reached 100 items or no more items to scrape.")
            break
        if not click_next_button():
            print("No more pages to scrape or error in navigating.")
            break
        time.sleep(5)  # Allow time for new page to load

# scraping
scrape_pages(3)  

# dataframe

df = pd.DataFrame({
    'Brands': brands,
    'Prices': prices,
    'Descriptions': descriptions
})
print(df)
df.to_csv("Flipkart Sneakers search.csv")
# driver.quit()
```

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementClickInterceptedException
import time

# Setup WebDriver
driver = webdriver.Chrome()
driver.get('https://www.amazon.in/')

# Search for 'Laptop'
search = driver.find_element(By.ID, 'twotabsearchtextbox')
search.send_keys('Laptop')
search.send_keys(Keys.ENTER)
time.sleep(5)

# Apply filter for 'Intel Core i7'
filters = driver.find_elements(By.CLASS_NAME, 'a-size-base.a-color-base')
for f in filters:
    if f.text == 'Intel Core i7':
        f.click()
        time.sleep(5)
        break

titles = []
ratings = []
prices = []

# Get all items
allitems = driver.find_elements(By.CLASS_NAME, 'puisg-col.puisg-col-4-of-12.puisg-col-8-of-16.puisg-col-12-of-20.puisg-col-12-of-24.puis-list-col-right')

# Limit to 10 items
item_limit = 10
item_count = 0

for item in allitems:
    if item_count >= item_limit:
        break
    
    try:
        # Extract title
        title = item.find_element(By.CLASS_NAME, 'a-section.a-spacing-none.puis-padding-right-small.s-title-instructions-style').text
        titles.append(title)
        
        # Extract rating
        try:
            rating_element = item.find_element(By.CLASS_NAME, 'a-icon-alt')
            rating_text = driver.execute_script("return arguments[0].innerText;", rating_element)
        except NoSuchElementException:
            rating_text = "No rating"
        ratings.append(rating_text)
        
        # Extract price
        try:
            price = item.find_element(By.CLASS_NAME, 'a-price-whole').text
        except NoSuchElementException:
            price = "No price"
        prices.append(price)
        
        item_count += 1  # Increment the item counter
        
    except StaleElementReferenceException:
        print("Stale element reference. Skipping item...")
        continue


# driver.quit()

# Create a DataFrame
df = pd.DataFrame({
    "Titles": titles,
    "Ratings": ratings,
    "Prices": prices
})

print(df)
df.to_csv("Amazon_laptop_search_results.csv", index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
from datetime import datetime
from dateutil.relativedelta import relativedelta


driver = webdriver.Chrome()


driver.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')
time.sleep(5) 


names = []
born_dates = []
terms_of_office = []


table = driver.find_element(By.XPATH, "//div[@class='TableData']//table")


rows = table.find_elements(By.TAG_NAME, "tr")[1:]  


def clean_date_string(date_str):
    
    if any(month in date_str for month in ['January', 'February', 'March', 'April', 'May', 'June',
                                           'July', 'August', 'September', 'October', 'November', 'December']):
        for month in ['January', 'February', 'March', 'April', 'May', 'June',
                      'July', 'August', 'September', 'October', 'November', 'December']:
            if month in date_str:
                date_str = date_str.replace(month, month + " ")
                break
    return date_str


def calculate_birth_date(age_str, start_date_str):

    age_parts = age_str.split(',')
    years = int(age_parts[0].split()[0])  
    days = int(age_parts[1].split()[0]) if len(age_parts) > 1 else 0  

    start_date_str = clean_date_string(start_date_str)
    
    start_date = datetime.strptime(start_date_str, '%d %B %Y')

    birth_date = start_date - relativedelta(years=years, days=days)
    
    return birth_date.strftime('%d %B %Y')

for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")
    
    name = cells[1].text
    names.append(name)
    
    term_start = cells[3].text
    terms_of_office.append(term_start)
    
    age_at_office = cells[2].text
    birth_date = calculate_birth_date(age_at_office, term_start)
    born_dates.append(birth_date)

# driver.quit()

# Create a DataFrame to hold the data
df = pd.DataFrame({
    "Name": names,
    "Born": born_dates,
    "Term Start": terms_of_office
})

# Save to a CSV file or display
df.to_csv("prime_ministers_data.csv", index=False)
print(df)
```

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time


driver = webdriver.Chrome()
driver.get('https://www.motor1.com/')


search = driver.find_element(By.CSS_SELECTOR, '.m1-search-panel-input.m1-search-form-text')
search.send_keys('50 most expensive cars')
search.send_keys(Keys.ENTER)

time.sleep(5)  

links = driver.find_elements(By.CSS_SELECTOR, '.text-box h3 a')
car_names=[]
prices=[]
for link in links:
    if '50 Most Expensive Cars in the World' in link.text:
        link.click()
        break


time.sleep(5)


all_items = driver.find_elements(By.CLASS_NAME, 'subheader')[:50]
# print(len(all_items))

for item in all_items:
    car_name = item.text
    car_names.append(car_name)
#     print(car_name)
    

    try:
        price = item.find_element(By.XPATH, 'following-sibling::p').text
#         print(price)
        prices.append(price)
    except Exception as e:
        print(f"Error finding price: {str(e)}")


# driver.quit()
df=pd.DataFrame({
    'Car Names':car_names,
    'Prices':prices
})
print(df)
df.to_csv('50 most expensive cars.csv')
```
