```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

keyword = input('Enter product to search on Amazon ')
driver = webdriver.Chrome()
driver.get('https://www.amazon.in/')


search = driver.find_element(By.CLASS_NAME, 'nav-input.nav-progressive-attribute')
search.send_keys(keyword, Keys.ENTER)


names = []
brands = []
prices = []
links = []
avs = []
dts = []
returns = []

def scrape_details():
    items = driver.find_elements(By.CLASS_NAME, 'puis-card-container.s-card-container.s-overflow-hidden.aok-relative.puis-expand-height.puis-include-content-margin.puis.puis-v37wyjnzl4u8m26eshvzoi5kqu.s-latency-cf-section.puis-card-border')
    
    for item in items:
        try:
            name = item.find_element(By.CLASS_NAME, 'a-size-base-plus.a-color-base.a-text-normal').text
        except:
            name = '-'
        names.append(name)

        brand = name.split()[0]
        brands.append(brand)

        try:
            price = item.find_element(By.CLASS_NAME, 'a-price-whole').text
        except:
            price = '-'
        prices.append(price)

        link_element = item.find_element(By.CLASS_NAME, 'a-link-normal.s-underline-text.s-underline-link-text.s-link-style.a-text-normal')
        link = link_element.get_attribute('href')
        links.append(link)

        try:
            availability = item.find_element(By.CLASS_NAME, 'a-size-base.a-color-price').text
        except:
            availability = 'In stock'
        avs.append(availability)

        try:
            delivery_time = item.find_element(By.CLASS_NAME, 'a-color-base.a-text-bold').text
        except:
            delivery_time = '-'
        dts.append(delivery_time)

def get_returns_policy(product_url):
    try:
        driver.get(product_url)
        

        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.ID, 'dp-container'))  
        )

        
        try:
            returns_policy_element = WebDriverWait(driver, 15).until(
                EC.presence_of_element_located((By.ID, 'RETURNS_POLICY'))
            )
            returns_policy = returns_policy_element.text
        except Exception as e:
            print(f"Returns policy element not found for {product_url}: {e}")
            returns_policy = 'No returns info found'
        
        return returns_policy
    except Exception as e:
        print(f"Error fetching returns policy for {product_url}: {e}")
        return 'No returns info found'


def next_page():
    try:
        
        nav_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CLASS_NAME, 's-pagination-item.s-pagination-next.s-pagination-button.s-pagination-separator'))
        )
        print("Next button found, clicking...")
        time.sleep(3)  

        
        nav_button.click()
        time.sleep(5)  
        return True
    except Exception as e:
        print(f"Error navigating to the next page: {e}")
        return False


for page in range(3):
    print(f"Scraping page {page + 1}...")
    scrape_details() 
    if page < 2:  
        if not next_page():
            break


for link in links:
    returns_info = get_returns_policy(link)
    returns.append(returns_info)


df = pd.DataFrame({
    "Brands": brands,
    "Names": names,
    "Prices": prices,
    "Expected Delivery": dts,
    "Availability": avs,
    "URLs": links,
    "Returns Policy": returns
})

print(df)
df.to_csv("Amazon_custom_search_with_returns.csv", index=False)
```

```python
import selenium
import re
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

keyword=input("Enter phone name")

driver=webdriver.Chrome()
driver.get('https://www.flipkart.in/')
search_bar=driver.find_element(By.CLASS_NAME,'Pke_EE')
search_bar.click()
search_bar.send_keys(keyword,Keys.ENTER)
time.sleep(5)
brands=[]
prices=[]
names=[]
urls=[]
colors=[]
dss=[]
storages=[]
pcs=[]
scs=[]
rams=[]
bcs=[]
global items,name
items=driver.find_elements(By.CLASS_NAME,'_75nlfW')
for item in items:
    name=item.find_element(By.CLASS_NAME,'KzDlHZ').text
    names.append(name)
    url_e=item.find_element(By.CLASS_NAME,'CGtC98')
    url=url_e.get_attribute('href')
    urls.append(url)
    brand=name.split()[0]
    brands.append(brand)
    price=item.find_element(By.CLASS_NAME,'Nx9bqj._4b5DiR').text
    prices.append(price)

for url in urls:
    driver.get(url)
    read_more_button = driver.find_element(By.CLASS_NAME, 'QqFHMw._4FgsLt')
    read_more_button.click()
    cols = driver.find_elements(By.CLASS_NAME, 'WJdYP6')
   

    primary_camera_found = False
    secondary_camera_found = False

    default_pc = '-'
    default_sc = '-'

    for c in cols:

        co = c.text
       
        
        if co.startswith('Color'):
            color = c.find_element(By.CLASS_NAME, 'HPETK2').text
            colors.append(color)
        
        elif re.match(r'^Display Size', co):
            display_size = c.find_element(By.CLASS_NAME, 'HPETK2').text
            dss.append(display_size)
     
        elif co.startswith('Internal'):
            storage = c.find_element(By.CLASS_NAME, 'HPETK2').text
            storages.append(storage)
      
        elif co.startswith('RAM'):
            ram = c.find_element(By.CLASS_NAME, 'HPETK2').text
            rams.append(ram)

        elif re.match(r'^Primary Camera\b', co):
            pc = c.find_element(By.CLASS_NAME, 'HPETK2').text
            primarycamerapattern = re.findall(r'\d+MP', pc)
            value = ' + '.join(primarycamerapattern)
            default_pc = value if value != '' else default_pc

        elif re.match(r'^Secondary Camera\b', co):
            sc = c.find_element(By.CLASS_NAME, 'HPETK2').text
            seccam = re.findall(r'(\d+MP|\d+\s*MP|\d+\.\dMP|\d+\.\d+\s*MP)', sc)
            value = ' + '.join(seccam)
            default_sc = value if value != '' else default_sc

        elif co.startswith('Battery') and co.split()[1] == 'Capacity':
            bat = c.find_element(By.CLASS_NAME, 'HPETK2').text
            bcs.append(bat)
        
    pcs.append(default_pc)
    scs.append(default_sc)




def fill_missing_values(*lists, fill_value='-'):
    max_len = max(len(lst) for lst in lists)
    for lst in lists:
        while len(lst) < max_len:
            lst.append(fill_value)


fill_missing_values(brands, names, colors, rams, storages, pcs, scs, dss, bcs, prices, urls)


df = pd.DataFrame({
    "Brand": brands,
    "Phone Name": names,
    "Color": colors,
    "Ram": rams,
    "Storage(ROM)": storages,
    "Primary Camera": pcs,
    "Secondary Camera": scs,
    "Display Size": dss,
    "Battery Capacity": bcs,
    "Price": prices,
    "URLs": urls
})

print(df)
df.to_csv('Flipkart_mobile_search.csv')
```
